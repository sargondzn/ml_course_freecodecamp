```python
import numpy as np
import matplotlib as plt
import pandas as pd
```

**Dataset**

Wolberg, W., Mangasarian, O., Street, N., & Street, W. (1993). Breast Cancer Wisconsin (Diagnostic) [Dataset]. UCI Machine Learning Repository. https://doi.org/10.24432/C5DW2B.

```python
#The columns are not labeled. So we should be able to find a way to label them.
cols = ["ID", "Diagnosis", "radius1", "texture1", "perimeter1", "area1", "smoothness1", "compactness1", "concavity1", "concave_points1", "symmetry1", "fractal_dimension1", "radius2", "texture2", "perimeter2", "area2", "smoothness2", "compactness2", "concavity2", "concave_points2", "symmetry2", "fractal_dimension2", "radius3", "texture3", "perimeter3", "area3", "smoothness3", "compactness3", "concavity3", "concave_points3", "symmetry3", "fractal_dimension3"]
df = pd.read_csv('wdbc.data', names = cols)
df.head()
```

```python
df["Diagnosis"] = (df["Diagnosis"]=="M").astype(int)
```

```python
df.head()
```

We want to predict wheter a diagnosis is going to be 'M' - malignant or "B' - benign

We are dealing with a **Supervised Leaning** kind of problem since the information we have is labeled and we are trying to predict something.

We are going to create a **Classification Model** since the output we want will be a categorical (qualitative) variable.

Feature: the columns that are in the dataset

Output: in this case, whether the tumor is benign or malignant

Samples: each line is a sample and each sample carries the different features (indepedent/input/entrance variables)

```python
#We want to create histograms that will help us understand how each feature's different numbers affect the probability of a cell's nuclei being benign or malignant.
#This way, we can compare some behaviours and try to understand a little more of the correlation between some aspects of our features.
#For example: if you analyse 'radious1', you will probably see that the bigger the nuclei gets, the bigger is the chance for it to be malignant


for label in cols[:-1]:
  plt.pyplot.hist(df[df["Diagnosis"]==1][label], color='blue', label='gamma', alpha=0.7, density=True)
  plt.pyplot.hist(df[df["Diagnosis"]==0][label], color='red', label='hadron', alpha=0.7, density=True)
  plt.pyplot.title(label)
  plt.pyplot.ylabel("Probability")
  plt.pyplot.xlabel(label)
  plt.pyplot.legend()
  plt.pyplot.show()

#Malignant = Blue
#Benign = Red
```
